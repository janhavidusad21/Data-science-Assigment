```python
for col in numeric_clos:
    Q1 = sales[col].quantile(0.25)
    Q3 = sales[col].quantile(0.75)
    IQR = Q3 - Q1
    print("IQR IS:",IQR)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = sales[(sales[col] < lower_bound) | (sales[col] > upper_bound)]

    print(f'\n{col} Outliers:')
    print(outliers)
```
